```python
sns.set(style = 'ticks' )
tips_df = pd.read_csv('tips.csv')
tips_df = tips_df.drop(['size'],axis = 1)
tips_df.head(20)
```

```python
g = sns.FacetGrid(tips_df , col ='time' , row = 'smoker')
```

```python
g = sns.FacetGrid(tips_df , row = 'time' , col = 'smoker')
g = g.map(plt.bar  , 'total_bill' , 'tip' )
```

```python
g = sns.FacetGrid(tips_df , row = 'smoker' , col = 'time')
g = g.map(plt.scatter , 'total_bill' , 'tip' ,edgecolor = 'w')
```

```python
g = sns.FacetGrid(tips_df , col = 'time' , hue = 'smoker')
g = g.map(plt.scatter , 'total_bill' , 'tip').add_legend()
```
